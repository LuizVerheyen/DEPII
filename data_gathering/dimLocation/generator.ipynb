```python
import pandas as pd
import numpy as np
import os
import glob

cwd = os.getcwd()
data_dir = os.path.join(cwd, "data")
if not os.path.exists(data_dir):
    data_dir = os.path.abspath(os.path.join(cwd, "..", "..", "data"))
```

```python
path_xls = os.path.join(data_dir, "zipcodes_num_nl_2025.xls")
path_zipcodes = os.path.join(data_dir, "zipcodes_num_nl_2025.csv")

if os.path.exists(path_xls) and not os.path.exists(path_zipcodes):
    temp_df = pd.read_excel(path_xls)
    temp_df.to_csv(path_zipcodes, index=False)
```

```python
path_staff = os.path.join(data_dir, "fietsvergoedingen_personeel_HOGENT_2024_2025.csv")
path_bluebike = os.path.join(data_dir, "bluebikes", "blue_bike_full_data.csv")
path_telpalen = os.path.join(data_dir, "telpalen", "GeoLocation", "telpalen_final_with_geoLocation.csv")

weather_station_pattern = os.path.join(data_dir, "weatherData", "aws_station", "aws_station_*.csv")
weather_station_files = glob.glob(weather_station_pattern)
path_weather_station = weather_station_files[0] if weather_station_files else None

zipcodes_df = pd.read_csv(path_zipcodes)
zipcodes_df.columns = zipcodes_df.columns.str.strip().str.lower()
staff_df = pd.read_csv(path_staff, sep=";")
blue_bike_df = pd.read_csv(path_bluebike, sep=";")
telpalen_df = pd.read_csv(path_telpalen, sep=";")
weather_station_df = pd.read_csv(path_weather_station, sep=",") if path_weather_station else pd.DataFrame()
```

```python
dim_location = zipcodes_df[["postcode", "plaatsnaam", "hoofdgemeente", "provincie"]].drop_duplicates().reset_index(drop=True)
dim_location.columns = ["PostalCode", "Municipality", "MainMunicipality", "Province"]
dim_location.insert(0, "LocationKey", range(1, len(dim_location) + 1))
dim_location = dim_location.assign(Latitude=np.nan, Longitude=np.nan, Address=np.nan)

loc_ref = dim_location[["LocationKey", "PostalCode", "Municipality", "Province"]]

dim_staff = staff_df.merge(loc_ref, left_on="Postcode of postcode groep", right_on="PostalCode", how="left") \
    [["PersoneelsID", "Entiteit", "Hoofdcampus", "LocationKey"]] \
    .rename(columns={"PersoneelsID": "StaffID", "Entiteit": "Entity", "Hoofdcampus": "Departure"})

dim_blue_bike = blue_bike_df.merge(loc_ref, on=["PostalCode", "Municipality"], how="left") \
    .assign(HasTrainConnection=1) \
    [["Location", "Address", "HasTrainConnection", "URL", "LocationKey"]]

dim_counting_point = telpalen_df.merge(loc_ref, left_on=["PostalCode", "Municipality"], right_on=["PostalCode", "Municipality"], how="left") 

station_mapping = {
    "BEITEM": "ROESELARE",
    "UCCLE": "UKKEL",
    "MONT RIGI": "WEISMES",
}
dim_weather_station = weather_station_df[['WeatherStationID', 'Name', 'Altitude', 'Longitude', 'Latitude', 'SnapshotDate']].copy()
dim_weather_station["MatchName"] = dim_weather_station["Name"].str.strip().str.upper().replace(station_mapping)
dim_loc_lookup = dim_location[["LocationKey", "Municipality"]].copy()
dim_loc_lookup["Municipality"] = dim_loc_lookup["Municipality"].str.strip().str.upper()
dim_weather_station = dim_weather_station.merge(
    dim_loc_lookup.drop_duplicates("Municipality"), 
    left_on="MatchName", 
    right_on="Municipality", 
    how="left"
).drop(columns=["MatchName", "Municipality"])
```

```python
output_dir = os.path.join(data_dir, "DimLocationData")
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

dim_location.to_csv(os.path.join(output_dir, "DimLocation.csv"), index=False)
dim_staff.to_csv(os.path.join(output_dir, "DimStaff.csv"), index=False)
dim_blue_bike.to_csv(os.path.join(output_dir, "DimBlueBikeStation.csv"), index=False)
dim_counting_point.to_csv(os.path.join(output_dir, "DimCountingPoint.csv"), index=False)
dim_weather_station.to_csv(os.path.join(output_dir, "DimWeatherStation.csv"), index=False)

print(f"Bestanden succesvol opgeslagen")
```
